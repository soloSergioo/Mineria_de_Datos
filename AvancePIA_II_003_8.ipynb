## Descripción de los datos
Son tablas que contienen estadísticas de audio de las 1994 pistas principales en Spotify. Los datos contienen aproximadamente 15 columnas, cada una de las cuales describe la pista y sus cualidades. Las canciones lanzadas desde 1956 hasta 2019 se incluyen de algunos artistas notables y famosos como Queen, The Beatles, Guns N 'Roses, etc. 

##### Variables
* **Index:** ID
* **Title:** Nombre de la pista
* **Artist:** Nombre del artista
* **Top Genre:** género de la pista
* **Year:** Año de lanzamiento de la pista
* **Beats per Minute (BPM):** el tempo de la canción
* **Energy:** la energía de una canción; cuanto mayor es el valor, más enérgico. canción
* **Danceability:** cuanto mayor sea el valor, más fácil será bailar esta canción.
* **Loudness:** cuanto mayor sea el valor, más fuerte será la canción.
* **Valence:** cuanto mayor sea el valor, más positivo será el estado de ánimo de la canción.
* **Length:** la duración de la canción en segundos.
* **Acousticness:** cuanto mayor sea el valor, más acústica será la canción.
* **Speechiness:** cuanto mayor sea el valor, más palabras habladas contiene la canción.
* **Popularity:** cuanto mayor sea el valor, más popular será la canción.

# Visualización  de los datos

```python
sns.pairplot(data_set[["Energy","Danceability","Valence","Loudness_(dB)","Acousticness"]].sample(1000))
```

Gráficos de dispersión e histogramas de los datos bivariantes más significativos.

**Energy vs Loudness_(dB)**

```python
sns.jointplot(x='Energy', y='Loudness_(dB)', data=data_set.sample(1000), kind='hex',gridsize=20)
```

Esta gráfica nos muestra que Energy y Loudness_(dB) están pobremente realcionados positivamente. Esto significa que cuanta mayor sea la energía más fuerte es la canción.

**Valence vs Danceability**

```python
sns.jointplot(x='Valence', y='Danceability', data=data_set.sample(100), kind='hex',gridsize=15)
```

Esta gráfica nos muestra que Valance y Danceability están pobremente realcionados positivamente. Esto significa que cuanto más positivo sea el estado de ánimo más fácil será bailar una canción.

**Energy vs Acousticness**

```python
sns.jointplot(x='Energy', y='Acousticness', data=data_set.sample(500), kind='hex',gridsize=20)
```

Esta gráfica nos muestra que Energy y Acousticness están pobremente realcionados negativamente. Esto significa que cuanto mayor es el valor enérgico menor acústica tiene la canción.
